```python
import io
import requests
```

```python
# Get the data
url = "https://raw.githubusercontent.com/ucla-econ-425t/2023winter/master/slides/data/Carseats.csv"
s = requests.get(url).content
Carseats = pd.read_csv(io.StringIO(s.decode('utf-8')))
Carseats
```

```python
# Numerical summaries
Carseats.info()
```

```python
# Describe the data
Carseats.describe()
```

```python
# Create a classification variable
Carseats["High"] = np.where(Carseats["Sales"] >= 8, 'Yes', 'No')
```

```python
# Split the data into training and testing set
from sklearn.model_selection import train_test_split

Carseats_train, Carseats_test = train_test_split(
  Carseats, 
  train_size = 0.75,
  random_state = 425, # seed
  )
print('Training set shape:', Carseats_train.shape)
print('Testing set shape:', Carseats_test.shape)
```

```python
cat_features = ['ShelveLoc', 'Urban', 'US']
cat_features
```

```python
num_features = Carseats_train.drop(['Sales','High', 'ShelveLoc', 'Urban', 'US'], axis = 1).columns
num_features
```

```python
features = np.concatenate([num_features, cat_features])
features
```

```python
# Separate X and y

# Training X and y
X_train = Carseats_train[features]
y_train = Carseats_train.High

# Testing X and y
X_test = Carseats_test[features]
y_test = Carseats_test.High
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Transform categorical predictors
categorical_tf = Pipeline(steps = [
  ("cat_impute", SimpleImputer(strategy = 'most_frequent')),
  ("encoder", OneHotEncoder(drop = 'first')),
  ("std", StandardScaler(with_mean = False))
])

# Transformer for continuous variables
numeric_tf = Pipeline(steps = [
  ("num_impute", SimpleImputer(strategy = 'mean')),
  ("std", StandardScaler())
])

# Column transformer
col_tf = ColumnTransformer(transformers = [
  ('num', numeric_tf, num_features),
  ('cat', categorical_tf, cat_features)
])
```

### SVM with linear kernel

```python
# Model
from sklearn.svm import SVC

svm_mod = SVC(
  C = 1.0,
  kernel = 'linear', 
  probability = True, 
  random_state = 425 
  )
```

```python
# Pipeline
from sklearn.pipeline import Pipeline

pipe = Pipeline(steps = [
  ("col_tf", col_tf),
  ("model", svm_mod)
  ])
pipe
```

```python
# Tune hyper-parameter(s)
C_grid = [0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5]
tuned_parameters = {
  "model__C": C_grid,
  }
tuned_parameters
```

```python
from sklearn.model_selection import GridSearchCV

# Set up CV
n_folds = 5
search = GridSearchCV(
  pipe,
  tuned_parameters, 
  cv = n_folds, 
  scoring = "roc_auc",
  # Refit the best model on the whole data set
  refit = True
  )
```

```python
# Fit CV
search.fit(X_train, y_train)
```

```python
# Visualize CV results
cv_res = pd.DataFrame({
  "C": np.array(search.cv_results_["param_model__C"]),
  "auc": search.cv_results_["mean_test_score"]
  })

plt.figure()
sns.relplot(
  # kind = "line",
  data = cv_res,
  x = "C",
  y = "auc"
  ).set(
    xlabel = "C",
    ylabel = "CV AUC"
    );
plt.show()
```

```python
# Best CV AUC
-search.best_score_
```

```python
# Training accuracy
from sklearn.metrics import accuracy_score, roc_auc_score

accuracy_score(
  y_train,
  search.best_estimator_.predict(X_train)
  )
```

```python
# Best estimator
search.best_estimator_
```

```python
# Final AUC on the test set
roc_auc_score(
  y_test,
  search.best_estimator_.predict_proba(X_test)[:, 1]
  )
```

```python
# Final classification accuracy
accuracy_score(
  y_test,
  search.best_estimator_.predict(X_test)
  )
```

$$\frac{1}{\|\beta\|} (\beta^T x - \beta^T x_0) = \frac{1}{\|\beta\|} (\beta^T x + \beta_0) = \frac{1}{\|\beta\|} f(x).$$
So $f(x)$ is proportional to the signed distance of a point $x$ to the hyperplane. It is $\|\beta\|$ times of the distance.
