### 1.1 Optimal regression function

$\begin{aligned}
\operatorname{E}[Y - \hat f(X)]^2 &= \operatorname{E}[f(X) + \varepsilon - \hat f(X)]^2 \\
&= \operatorname{E}[(f(X) - \hat f(X))^2] + \operatorname{E}[\varepsilon^2] + 2\operatorname{E}[\varepsilon(f(X) - \hat f(X))]
\end{aligned}$

Because: \
$\begin{aligned}
\operatorname{E}[\varepsilon(f(X) - \hat f(X))] &=
\operatorname{E}[\operatorname{E}[\varepsilon(f(X) - \hat f(X))]|X] \\
&= \operatorname{E}[\operatorname{E}[\varepsilon|X](f(X) - \hat f(X))] \\
&= 0
\end{aligned}$

And $\operatorname{E}[\varepsilon^2]$ is a constant number.

When $\hat f(X) = \operatorname{E}[Y|X] = \operatorname{E}[f(X) + \varepsilon|X] = f(X)$, $\operatorname{E}[(f(X) - \hat f(X))^2]$ arrives the minimum value $0$ and mean squared prediction error arrives the minimum value $\operatorname{E}[\varepsilon^2]$.

### 1.2 Bias-variance trade-off

$\begin{aligned}
\operatorname{E}[y_0 - \hat f(x_0)]^2 &= \operatorname{E}[f(x_0) + \varepsilon - \hat f(x_0)]^2 \\
&= \operatorname{E}[(f(x_0) - \hat f(x_0))^2] + \operatorname{E}[\varepsilon^2] + 2\operatorname{E}[\varepsilon(f(x_0) - \hat f(x_0))] \\
&= \operatorname{E}[(f(x_0) - \hat f(x_0))^2] + \operatorname{Var}[\varepsilon] + 0
\end{aligned}$

Because: \
$\begin{aligned}
\operatorname{E}[(f(x_0) - \hat f(x_0))^2] &= \operatorname{E}[(f(x_0) - \operatorname{E}[\hat f(x_0)] + \operatorname{E}[\hat f(x_0)] - \hat f(x_0))^2] \\
&= \operatorname{E}[(f(x_0) - \operatorname{E}[\hat f(x_0)])^2] + \operatorname{E}[(\operatorname{E}[\hat f(x_0)] - \hat f(x_0))^2] + \\
&\operatorname{E}[(f(x_0) - \operatorname{E}[\hat f(x_0)])(\operatorname{E}[\hat f(x_0)] - \hat f(x_0))] \\
& = \operatorname{E}[(f(x_0) - \operatorname{E}[\hat f(x_0)])^2] + \operatorname{E}[(\operatorname{E}[\hat f(x_0)] - \hat f(x_0))^2] + 0\\
&= [\operatorname{Bias}(\hat f(x_0))]^2 + \operatorname{Var}(\hat f(x_0))
\end{aligned}$

We can get: \
$\operatorname{E}[y_0 - \hat f(x_0)]^2 = \operatorname{Var}(\hat f(x_0)) + [\operatorname{Bias}(\hat f(x_0))]^2 + \operatorname{Var}(\varepsilon)$

## 6. ISL Exercise 3.7.15

### (a)

```python
import statsmodels.formula.api as smf

columns = Boston.columns.difference(['crim'])
slr_coeff = [
    smf.ols(formula = 'crim ~ ' + col, data = Boston).fit().params[1]
    for col in columns
    ]
slr_se = [
    smf.ols(formula = 'crim ~ ' + col, data = Boston).fit().bse[1]
    for col in columns
    ]
slr_t = [
    smf.ols(formula = 'crim ~ ' + col, data = Boston).fit().tvalues[1]
    for col in columns
    ]
slr_p = [
    smf.ols(formula = 'crim ~ ' + col, data = Boston).fit().pvalues[1]
    for col in columns
    ]
slr_result = pd.DataFrame({'Betahat': slr_coeff, 'std error': slr_se,
                          't': slr_t, 'p-value': slr_p}, index = columns)
slr_result
```

Because all the p-values of the simple linear regression coefficient except for $chas$ are smaller than 0.05, there is a statistically significant association between the predictors except for $chas$ and the response.

```python
fig, axes = plt.subplots(3, 4, figsize = (20, 15),sharey=True)
for i in range(3):
    for j in range(4):
        sns.regplot(x=Boston[columns[(i)*4+j]],y=Boston['crim'], 
                    scatter_kws = {'alpha' : 0.3},
                    line_kws = {"color":"orange"},lowess = True,ax = axes[i,j])
```

From the scatter plots and lowess lines, we can see there is a statistically significant association between the predictors except for $chas$ and the response.

### (b)

```python
mlr_reg = smf.ols(formula = 'crim ~ ' + '+'.join(columns), data = Boston).fit()
mlr_result = mlr_reg.summary()
mlr_result
```

The result is shown on the table above and it shows that strong multicollinearity problem exists. For $dis$, $medv$, $rad$ and $zn$ we can reject the null hypothesis $H_0 : Î²_j = 0$, because their p-value is less than 0.05.

```python
nlr_coeff1 = [
    smf.ols(formula = 'crim ~ ' + col + '+ I(' +col+ '**2)' + '+ I(' +col+ '**3)', 
            data = Boston).fit().params[1]
    for col in columns
    ]
nlr_coeff2 = [
    smf.ols(formula = 'crim ~ ' + col + '+ I(' +col+ '**2)' + '+ I(' +col+ '**3)', 
            data = Boston).fit().params[2]
    for col in columns
    ]
nlr_coeff3 = [
    smf.ols(formula = 'crim ~ ' + col + '+ I(' +col+ '**2)' + '+ I(' +col+ '**3)', 
            data = Boston).fit().params[3]
    for col in columns
    ]
nlr_p1 = [
    smf.ols(formula = 'crim ~ ' + col + '+ I(' +col+ '**2)' + '+ I(' +col+ '**3)', 
            data = Boston).fit().pvalues[1]
    for col in columns
    ]
nlr_p2 = [
    smf.ols(formula = 'crim ~ ' + col + '+ I(' +col+ '**2)' + '+ I(' +col+ '**3)', 
            data = Boston).fit().pvalues[2]
    for col in columns
    ]
nlr_p3 = [
    smf.ols(formula = 'crim ~ ' + col + '+ I(' +col+ '**2)' + '+ I(' +col+ '**3)', 
            data = Boston).fit().pvalues[3]
    for col in columns
    ]
nlr_result = pd.DataFrame({'Betahat1': nlr_coeff1, 'p-value1': nlr_p1, 
                           'Betahat2': nlr_coeff2, 'p-value2': nlr_p2,
                           'Betahat3': nlr_coeff3, 'p-value3': nlr_p3}, index = columns)
nlr_result
```

The p-value of the coefficients for the square of the predictors and the predictors to the power of three are all significant for $age$, $dis$, $indus$, $medv$, $nox$ and $ptratio$. So there is evidence of non-linear association between the response and these predictors.

$\begin{aligned}
R^2 &= 1 - \frac{\text{RSS}}{\text{TSS}} = \frac{\text{ESS}}{\text{TSS}} = \frac{\operatorname{Var}[\hat{\mathbf{y}}]}{\operatorname{Var}[\mathbf{y}]} \\
[\operatorname{Cor}(\mathbf{y}, \hat{\mathbf{y}})]^2 &= \frac{\operatorname{Cov}(\mathbf{y}, \hat{\mathbf{y}})^2}{\operatorname{Var}[\mathbf{y}]\operatorname{Var}[\hat{\mathbf{y}}]} \\
\operatorname{Cov}(\mathbf{y}, \hat{\mathbf{y}}) &= \operatorname{Cov}(\hat{\mathbf{y}}, \hat{\mathbf{y}} + \mathbf{\hat{u}}) \\
&= \operatorname{Cov}(\hat{\mathbf{y}}, \hat{\mathbf{y}}) + \operatorname{Cov}(\hat{\mathbf{y}}, \mathbf{\hat{u}}) \\
&= \operatorname{Cov}(\hat{\mathbf{y}}, \hat{\mathbf{y}}) + 0 \\
&= \operatorname{Var}(\hat{\mathbf{y}}) \\
[\operatorname{Cor}(\mathbf{y}, \hat{\mathbf{y}})]^2 &= \frac{\operatorname{Var}[\hat{\mathbf{y}}]}{\operatorname{Var}[\mathbf{y}]}
\end{aligned}$

So $R^2 = [\operatorname{Cor}(\mathbf{y}, \hat{\mathbf{y}})]^2$.
